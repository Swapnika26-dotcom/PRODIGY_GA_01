```python
from IPython.display import HTML, display
import random
```

```python
def markov_like(text):
    words = text.split()
    if len(words) < 2:
        return text

    chain = {}
    for i in range(len(words) - 1):
        chain.setdefault(words[i], []).append(words[i+1])

    start = random.choice(words)
    result = [start]

    for _ in range(15):
        next_words = chain.get(result[-1])
        if not next_words:
            break
        result.append(random.choice(next_words))

    return " ".join(result)
```

```python
def show_app():
    html = """
    <h2>Markov Text Generator</h2>
    <input id="text" style="width:400px">
    <button onclick="generate()">Generate</button>
    <p id="output"></p>

    <script>
    function generate(){
        var txt = document.getElementById("text").value;
        google.colab.kernel.invokeFunction(
            'notebook.generate',
            [txt],
            {}
        );
    }
    </script>
    """
    display(HTML(html))
```

```python
from google.colab import output

def generate(text):
    result = markov_like(text)
    display(HTML(f"<b>Generated Output:</b><br>{result}"))

output.register_callback('notebook.generate', generate)
```

```python
show_app()
```
